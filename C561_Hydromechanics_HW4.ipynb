## 0.Linear Function and derivatives

Simplified form of the linear velocity potential function is would be like that:


```python
%%latex
\dfrac{Hg\cosh\left(k\left(z+d\right)\right)\sin\left({\omega}t-kx\right)}{2{\omega}\cosh\left(kd\right)} = {\phi}_\text{linear}
```

If we put the known values (like k,H,d and w) on the equation, it would be:

```python
%%latex
{\phi}_\text{linear} = -\dfrac{24525\cosh\left(\frac{3\left(z+8\right)}{25}\right)\sin\left(\frac{3x}{25}-\frac{469t}{500}\right)}{1876\cosh\left(\frac{24}{25}\right)}
```

Derivative of phi in (x direction)

Derivative of the function gives the velocity components.

```python
%%latex
\frac {\partial \phi}{ \partial x} = u
```

```python
%%latex
u = -\dfrac{2943\cosh\left(\frac{3\left(z+8\right)}{25}\right)\cos\left(\frac{60x-469t}{500}\right)}{1876\cosh\left(\frac{24}{25}\right)}
```

Derivative of phi (z direction)

```python
%%latex
\frac {\partial \phi}{ \partial z} = w \\
w = -\dfrac{2943\sin\left(\frac{3x}{25}-\frac{469t}{500}\right)\sinh\left(\frac{3\left(z+8\right)}{25}\right)}{1876\cosh\left(\frac{24}{25}\right)}
```

## 0.Stokes Function and derivatives

```python
%%latex
\dfrac{Hg\cosh\left(k\left(h+z\right)\right)\sin\left({\omega}t-kx\right)}{2{\omega}\cosh\left(kd\right)} -\dfrac{3H^2{\omega}\cosh\left(2k\left(z+h\right)\right)\sin\left(2\left(kx-{\omega}t\right)\right)}{32\sinh^4\left(kh\right)} = {\phi}_\text{stokes2}
```

If we put the known values the equation would be:

```python
%%latex
{\phi}_\text{stokes2} = -\dfrac{1407\cosh\left(\frac{6\left(z+8\right)}{25}\right)\sin\left(2\left(\frac{3x}{25}-\frac{469t}{500}\right)\right)}{2560\sinh^4\left(\frac{24}{25}\right)}-\dfrac{460089\cosh\left(\frac{3\left(z+8\right)}{25}\right)\sin\left(\frac{3x}{25}-\frac{469t}{500}\right)}{40000\cosh\left(\frac{24}{25}\right)}
```

Derivative of phi in (x direction)

Derivative of the function gives the velocity components.

```python
%%latex
\frac {\partial \{{\phi}_{stokes}}{ \partial x} = u
```

```python
%%latex
u = -\dfrac{4221\cosh\left(\frac{6\left(z+8\right)}{25}\right)\cos\left(2\left(\frac{3x}{25}-\frac{469t}{500}\right)\right)}{32000\sinh^4\left(\frac{24}{25}\right)}-\dfrac{1380267\cosh\left(\frac{3\left(z+8\right)}{25}\right)\cos\left(\frac{3x}{25}-\frac{469t}{500}\right)}{1000000\cosh\left(\frac{24}{25}\right)}
```

```python
%%latex
\frac {\partial \{{\phi}_{stokes}}{ \partial z} = w
```

## 6. Stream Function

Plot the streamlines using appropriate intervals for stream function Ïˆ

```python
# Streamline equation
# (24525*sinh((3*(z+8))/25)*cos((3*x)/25-(469*t)/500))/(1876*cosh(24/25))
```

```python
x = np.arange(0,10,0.1)
z = np.arange(0,10,0.1)
t = np.zeros(100) # To eliminate time variable
STR = np.arange(0,10,0.1)
for i in range(len(x)):
  STR[i] = (24525*math.sinh((3*(z[i]+8))/25)*math.cos((3*x[i])/25-(469*t[i])/500))/(1876*math.cosh(24/25))
# STR is the stream function
```

```python
import plotly.figure_factory as ff

# Source: https://plotly.com/python/streamline-plots/
Y, X = np.meshgrid(x, STR)
u = -1 - X**2 + Y
w = 1 + X - Y**2

# Create streamline figure
fig = ff.create_streamline(x, STR, u, w, arrow_scale=.1)
fig.show()
```
